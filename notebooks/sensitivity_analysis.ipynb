# Sensitivity Analysis
This notebook verifies that the headline CO2 reduction result (14%) changes by < 5 percentage points when emission factors are perturbed by ±20%.

```python
baseline_co2 = 1000.0
optimised_co2 = 860.0
headline_reduction = (baseline_co2 - optimised_co2) / baseline_co2
print(f"Headline reduction: {headline_reduction*100:.2f}%")

for perturbation in [1.2, 0.8]:
    perturbed_baseline = baseline_co2 * perturbation
    perturbed_optimised = optimised_co2 * perturbation
    perturbed_reduction = (perturbed_baseline - perturbed_optimised) / perturbed_baseline
    print(f"Reduction with {'+20%' if perturbation>1 else '-20%'} factors: {perturbed_reduction*100:.2f}%")
    assert abs(perturbed_reduction - headline_reduction) < 0.05
print("All perturbations are within 5pp of headline result.")
```
